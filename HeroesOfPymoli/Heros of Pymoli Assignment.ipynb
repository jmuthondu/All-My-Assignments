```python
# Read the File and store into Pandas data frame
Pymoli_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\Pymoli_data.csv")
```

```python
#Using the length of the list in 'SN' to find the total number of players
total_players =len(Pymoli_df["SN"].value_counts())

# Creating a dataframe with the total number of players
player_count = pd.DataFrame({"Total Players": [total_players]})
player_count
```

**Purchasing Analysis (Total)**

- Run basic calculations to obtain number of unique items, average price, etc.
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame

```python
#Calculation for unique items, the average price,number of purchases and the total revenue
number_of_unique_items = len((Pymoli_df["Item ID"]).unique()) 
average_price = Pymoli_df["Price"].mean()
number_of_purchases = Pymoli_df["Purchase ID"].count()
total_revenue = Pymoli_df["Price"].sum()

#Creating a DataFrame from the data as obtained above
summary_df = pd.DataFrame.from_dict({"Unique Items": [number_of_unique_items],
                           "Average Price": [average_price],
                           "Purchases": [number_of_purchases],
                           "Total Revenue": [total_revenue]}, orient ="columns")
summary_df
```

```python
# A Display summary of the dataframe with a cleaner format by adding a dollar sign on the average price and the total revenue
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

summary_df
          
```

**Gender Demographics**

- Percentage and Count of Male Players
- Percentage and Count of Female Players
- Percentage and Count of Other / Non-Disclosed

```python
# Grouping by gender & counting
gender = Pymoli_df[["SN", "Gender"]]

gender = gender.drop_duplicates()

counts = gender["Gender"].value_counts()

# List of values
total_counts = [counts[0],counts[1],counts[2]]
percents = [round((counts[0]/total_players)*100,2),round((counts[1]/total_players)*100,2),round((counts[2]/total_players)*100,2)]

# Creating DataFrame & setting index
gender_demo = pd.DataFrame({ "Percentage of Players": percents,"Total Count": total_counts
})
gender_demo.index = (["Male", "Female", "Other / Non-Disclosed"])
gender_demo
```

**Purchasing Analysis (Gender)**

- Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame

```python
#Purchasing Analysis (Gender)
purchase_count=Pymoli_df.groupby(["Gender"]).count()["Price"]
purchase_avg_price=Pymoli_df.groupby(["Gender"]).mean()["Price"]
total_purchase_value=Pymoli_df.groupby(["Gender"]).sum()["Price"]
normalized_totals=total_purchase_value/counts

summary_purchasing_analysis=pd.DataFrame({"Purchase Count":purchase_count,
                                         "Average Purchase Price":purchase_avg_price,
                                         "Total Purchase Value": total_purchase_value,
                                         "Normalized Total":normalized_totals})

summary_purchasing_analysis=summary_purchasing_analysis[["Purchase Count","Average Purchase Price","Total Purchase Value","Normalized Total"]]

summary_purchasing_analysis=summary_purchasing_analysis.round(2)
summary_purchasing_analysis["Average Purchase Price"] = summary_purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
summary_purchasing_analysis["Total Purchase Value"] = summary_purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
summary_purchasing_analysis
```

**Age Demographics**

- Establish bins for ages
- Categorize the existing players using the age bins. Hint: use pd.cut()
- Calculate the numbers and percentages by age group
- Create a summary data frame to hold the results
- Optional: round the percentage column to two decimal points
- Display Age Demographics Table

```python
#Establish bins for ages
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_df = Pymoli_df.loc[:,:]

age_df["Age Group"] = pd.cut(age_df["Age"], age_bins, labels=group_names)

# Create a GroupBy object based upon "Age Group"
age_group = age_df.groupby("Age Group")

age_demo = age_group.first()
age_demo = age_demo.drop("Purchase ID",1)
age_demo = age_demo.drop("SN",1)
age_demo = age_demo.drop("Age",1)
age_demo = age_demo.drop("Gender",1)
age_demo = age_demo.drop("Item ID",1)
age_demo = age_demo.drop("Item Name",1)
age_demo = age_demo.drop("Price",1)

age_num = age_group[["SN"]].nunique()
age_demo["Total Count"] = age_num

age_demo["Percentage of Players"] = 100 * age_num / total_players
age_demo["Percentage of Players"] = age_demo["Percentage of Players"].map("{:,.2f}".format)
age_demo
```

**Purchasing Analysis (Age)**

- Bin the purchase\_data data frame by age
- Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame

```python
#Calculations for purchase count,average purchase price etc
age_purchase_count= Pymoli_df.groupby("Age Group")["Purchase ID"].count()
total_purchase_count = Pymoli_df.groupby("Age Group")["SN"].nunique()
avg_purchase_price = Pymoli_df.groupby("Age Group")["Price"].mean()
total_purchase_value = Pymoli_df.groupby("Age Group")["Price"].sum()

#Summary dataframe to hold the results with the data being formatted

age_purchasing_analysis = pd.DataFrame({
    "Purchase Count": age_purchase_count,
    "Average Purchase Price": round(avg_purchase_price,2),
    "Total Purchase Value": total_purchase_value,
    "Avg Total Purchase per Person": round(total_purchase_value/total_purchase_count,2)
})
age_purchasing_analysis["Average Purchase Price"] = age_purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
age_purchasing_analysis["Total Purchase Value"] = age_purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
age_purchasing_analysis["Avg Total Purchase per Person"] = age_purchasing_analysis["Avg Total Purchase per Person"].map("${:,.2f}".format)

age_purchasing_analysis
```

**Top Spenders**

- Run basic calculations to obtain the results in the table below
- Create a summary data frame to hold the results
- Sort the total purchase value column in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the summary data frame

```python
user_purchase_count= Pymoli_df.groupby("SN")["Price"].count()
user_avg_purchase_price = Pymoli_df.groupby("SN")["Price"].mean()
user_total_purchase_value = Pymoli_df.groupby("SN")["Price"].sum()

#Summary of the dataframe holding the results with a cleaner format
top_spenders_df = pd.DataFrame({
    "Purchase Count": user_purchase_count,
    "Average Purchase Price": round(user_avg_purchase_price,2),
    "Total Purchase Value": user_total_purchase_value
})
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)
#Sorting the total purchase column in a descending order
top_spenders_df.sort_values("Total Purchase Value", ascending=False).head()
```

**Most Popular Items**

- Retrieve the Item ID, Item Name, and Item Price columns
- Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
- Create a summary data frame to hold the results
- Sort the purchase count column in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the summary data frame

```python
#Retrieving Item ID, Item Name and the Item Price columns
retrieved_items = Pymoli_df[["Item ID", "Item Name","Price"]]
retrieved_items.head()                           
```

```python
#Grouping by Item ID and Item Name and performing the required calculations for purchase count
purchase_count = Pymoli_df.groupby(["Item ID", "Item Name"])["Price"].count()

#Grouping by Item ID and Item Name and performing the required calculations for item price
item_price = Pymoli_df.groupby(["Item ID", "Item Name"])["Price"].mean()

#Grouping by Item ID and Item Name and performing the required calculations for total purchase value
total_purchases = Pymoli_df.groupby(["Item ID", "Item Name"])["Price"].sum()

#Summary dataframe to hold the results with a clean format

popular_summary_df = pd.DataFrame({"Purchase Count":purchase_count,
                                   "Item Price":item_price,
                                   "Total Purchase Value":total_purchases})
popular_summary_df["Item Price"] = popular_summary_df["Item Price"].map("${:,.2f}".format)
popular_summary_df["Total Purchase Value"] = popular_summary_df["Total Purchase Value"].map("${:,.2f}".format)

popular_summary_df.sort_values(by="Purchase Count",ascending= False).head()
```
